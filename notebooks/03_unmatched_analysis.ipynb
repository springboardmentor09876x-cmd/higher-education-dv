```python
unmatched = pd.read_csv("../data/processed/openalex_unmatched.csv")

print("Rows :", len(unmatched))

unmatched.head()
```

```python
unmatched["university_name"].sample(100, random_state=42)
```

```python
keywords = [
    "technology",
    "institute",
    "college",
    "science",
    "medical",
    "engineering",
    "business",
    "management",
    "polytechnic",
    "university",
]

for word in keywords:

    count = unmatched["university_name"].str.contains(
        word,
        case=False,
        na=False,
    ).sum()

    print(f"{word:<15} {count}")
```

```python
unmatched[
    unmatched["university_name"]
    .str.contains(
        r"\(",
        regex=True,
        na=False,
    )
][
    [
        "university_name",
        "country",
    ]
]
```

```python
unmatched[
    unmatched["university_name"]
    .str.len() < 30
][
    [
        "university_name",
        "country",
    ]
].head(100)
```
