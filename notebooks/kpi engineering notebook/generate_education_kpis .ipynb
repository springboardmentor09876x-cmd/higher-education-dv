## Import Libraries

```python
import pandas as pd

df = pd.read_csv(
    r"C:\Users\s sai vaishnavi\OneDrive\Desktop\Higher_Education_Project\datasets\processed_data\cleaned data\final_cleaned_university_dataset.xls"
)
```

## Load Cleaned Dataset

```python
df["world_rank_numeric"] = (
    df["world_rank"]
      .astype(str)
      .str.replace("=", "", regex=False)
      .str.replace("+", "", regex=False)
      .str.extract(r"(\d+)", expand=False)
)

df["world_rank_numeric"] = pd.to_numeric(
    df["world_rank_numeric"],
    errors="coerce"
)
```

## Explore Dataset

```python
df["global_ranking_score"] = (
    (
        df["world_rank_numeric"].max() - df["world_rank_numeric"]
    )
    /
    (
        df["world_rank_numeric"].max() - df["world_rank_numeric"].min()
    )
    * 100
).round(2)
```

```python
df["research_impact_score"] = (
    (
        df["citations_score"] +
        df["citations_per_faculty"]
    ) / 2
).round(2)
```

```python
df["faculty_to_student_kpi"] = df["faculty_to_student_ratio"]
```
