```python
qs_history = pd.read_csv(RAW_DATA / "qs_rankings_2017_2022_raw.csv")

qs_2024 = pd.read_csv(RAW_DATA / "qs_rankings_2024_raw.csv")

the_2020 = pd.read_csv(RAW_DATA / "2020_rankings.csv")
the_2021 = pd.read_csv(RAW_DATA / "2021_rankings.csv")
the_2022 = pd.read_csv(RAW_DATA / "2022_rankings.csv")
the_2023 = pd.read_csv(RAW_DATA / "2023_rankings.csv")
the_2024 = pd.read_csv(RAW_DATA / "2024_rankings.csv")
```

```python
def audit_dataset(df, dataset_name):

    print("=" * 80)
    print(dataset_name)
    print("=" * 80)

    print(f"\nShape : {df.shape}")

    print("\nColumns")
    print(df.columns.tolist())

    print("\nData Types")
    print(df.dtypes)

    print("\nMissing Values")
    print(df.isnull().sum())

    print("\nDuplicate Rows")
    print(df.duplicated().sum())

    print("\nFirst Five Rows")
    display(df.head())
```

```python
audit_dataset(qs_history, "QS 2017-2022")

audit_dataset(qs_2024, "QS 2024")

audit_dataset(the_2020, "THE 2020")

audit_dataset(the_2021, "THE 2021")

audit_dataset(the_2022, "THE 2022")

audit_dataset(the_2023, "THE 2023")

audit_dataset(the_2024, "THE 2024")
```

```python
import pandas as pd

# Load datasets
registry = pd.read_csv("../data/interim/registry/university_registry.csv")
openalex = pd.read_csv("../data/raw/openalex_bulk_raw.csv")

# Normalize names in the same way
registry_names = registry["normalized_name"]

openalex_names = (
    openalex["openalex_display_name"]
    .fillna("")
    .str.lower()
    .str.strip()
)

# Count exact matches
matched = registry_names.isin(openalex_names)

print("=" * 50)
print("Registry Universities :", len(registry))
print("Exact Matches         :", matched.sum())
print("Coverage              :",
      round(matched.sum() / len(registry) * 100, 2), "%")
print("=" * 50)
```
