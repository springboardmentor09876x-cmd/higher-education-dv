```python
import pandas as pd

# Load files
master = pd.read_csv("master_common_universities.csv")
qs = pd.read_csv("qs2026_clean.csv")
the = pd.read_csv("the_clean.csv")
cwur = pd.read_csv("cwur_clean.csv")

# Standardize names
for df in [master, qs, the, cwur]:
    df["clean_name"] = (
        df["University_Name"]
        .astype(str)
        .str.lower()
        .str.strip()
    )

# Merge QS
merged = master.merge(
    qs,
    on="clean_name",
    how="left",
    suffixes=("", "_qs")
)

# Merge THE
merged = merged.merge(
    the,
    on="clean_name",
    how="left",
    suffixes=("", "_the")
)

# Merge CWUR
merged = merged.merge(
    cwur,
    on="clean_name",
    how="left",
    suffixes=("", "_cwur")
)

print("Rows:", merged.shape[0])
print("Columns:", merged.shape[1])

merged.to_csv(
    "master_university_dataset.csv",
    index=False
)

print("master_university_dataset.csv created successfully!")
```

```python
import pandas as pd

# Load files
master = pd.read_csv("master_university_dataset.csv")
the2023 = pd.read_csv("the2023_clean.csv")

# Create clean names
master["clean_name"] = (
    master["University_Name"]
    .astype(str)
    .str.lower()
    .str.strip()
)

the2023["clean_name"] = (
    the2023["University_Name"]
    .astype(str)
    .str.lower()
    .str.strip()
)

# Merge
master_final = master.merge(
    the2023,
    on="clean_name",
    how="left",
    suffixes=("", "_the2023")
)

print("Rows:", master_final.shape[0])
print("Columns:", master_final.shape[1])

master_final.to_csv(
    "master_dataset_with_students.csv",
    index=False
)

print("master_dataset_with_students.csv created successfully!")
```

```python
import pandas as pd
import requests
import time
from tqdm import tqdm

universities = pd.read_csv("openalex_university_list.csv")

results = []

for uni in tqdm(universities["University_Name"]):

    try:
        url = f"https://api.openalex.org/institutions?search={uni}"

        response = requests.get(url, timeout=20)

        if response.status_code != 200:
            continue

        data = response.json()

        if len(data["results"]) == 0:
            continue

        inst = data["results"][0]

        subject_field = None

        if "topics" in inst and len(inst["topics"]) > 0:
            subject_field = inst["topics"][0].get("display_name")

        results.append({
            "University_Name": uni,
            "publications_count": inst.get("works_count"),
            "citations_count": inst.get("cited_by_count"),
            "h_index": inst.get("summary_stats", {}).get("h_index"),
            "i10_index": inst.get("summary_stats", {}).get("i10_index"),
            "subject_field": subject_field
        })

        time.sleep(0.1)

    except Exception:
        continue

openalex_df = pd.DataFrame(results)

print("Records fetched:", len(openalex_df))

openalex_df.to_csv(
    "openalex_research_metrics.csv",
    index=False
)

print("Saved: openalex_research_metrics.csv")
```

```python
import pandas as pd
import requests
import time
from tqdm import tqdm

universities = pd.read_csv("openalex_university_list_v2.csv")

results = []

for _, row in tqdm(universities.iterrows(),
                   total=len(universities)):

    original_name = row["University_Name"]
    search_name = row["search_name"]

    try:

        url = (
            "https://api.openalex.org/institutions"
            f"?search={search_name}"
        )

        response = requests.get(url, timeout=20)

        if response.status_code != 200:
            continue

        data = response.json()

        if len(data["results"]) == 0:
            continue

        inst = data["results"][0]

        subject_field = None

        if (
            "topics" in inst
            and len(inst["topics"]) > 0
        ):
            subject_field = (
                inst["topics"][0]
                .get("display_name")
            )

        results.append({
            "University_Name": original_name,
            "Matched_Name": inst.get("display_name"),
            "publications_count": inst.get("works_count"),
            "citations_count": inst.get("cited_by_count"),
            "h_index": inst.get("summary_stats", {})
                            .get("h_index"),
            "i10_index": inst.get("summary_stats", {})
                             .get("i10_index"),
            "subject_field": subject_field
        })

        time.sleep(0.1)

    except Exception:
        continue

openalex_df_v2 = pd.DataFrame(results)

print("Records fetched:",
      len(openalex_df_v2))

openalex_df_v2.to_csv(
    "openalex_research_metrics_v2.csv",
    index=False
)

print("Saved successfully!")
```

```python
import numpy as np

final_eduvision = master_final.copy()

# Research Excellence Index
final_eduvision["Research_Excellence_Index"] = (
    0.4 * final_eduvision["Research_Score"].fillna(0) +
    0.3 * final_eduvision["Citation_Score"].fillna(0) +
    0.2 * final_eduvision["Publications"].fillna(0) +
    0.1 * final_eduvision["Influence"].fillna(0)
)

# Innovation Index
final_eduvision["Innovation_Index"] = (
    final_eduvision["Patents"].fillna(0) +
    final_eduvision["Influence"].fillna(0)
)

# Student Experience Index
final_eduvision["Student_Experience_Index"] = (
    0.4 * final_eduvision["Teaching_Score_2023"].fillna(0) +
    0.3 * final_eduvision["International_Student_Ratio_2023"].fillna(0) +
    0.3 * (100 - final_eduvision["Faculty_Student_Ratio_2023"].fillna(0))
)

# Global Reputation Index
final_eduvision["Global_Reputation_Index"] = (
    0.6 * final_eduvision["Academic_Reputation"].fillna(0) +
    0.4 * final_eduvision["Employer_Reputation"].fillna(0)
)

# Research Productivity Score
final_eduvision["Research_Productivity_Score"] = (
    final_eduvision["Publications"].fillna(0) +
    final_eduvision["Citation_Score"].fillna(0)
)

# Internationalization Score
final_eduvision["Internationalization_Score"] = (
    final_eduvision["International_Students_Ratio"].fillna(0) +
    final_eduvision["International_Faculty_Ratio"].fillna(0) +
    final_eduvision["International_Outlook"].fillna(0)
) / 3

print("New Shape:", final_eduvision.shape)
```
