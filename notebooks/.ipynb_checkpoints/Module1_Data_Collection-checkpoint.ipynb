```python
import pandas as pd
from pathlib import Path

# ==========================================
# Module 1 - Data Collection
# Create Raw Master Dataset
# ==========================================

RAW_PATH = Path("../datasets/raw")
OUTPUT_PATH = Path("../datasets/final")
OUTPUT_PATH.mkdir(parents=True, exist_ok=True)

all_datasets = []

encodings = ["utf-8", "latin1", "cp1252"]

for file in RAW_PATH.rglob("*.csv"):

    print(f"Reading {file.name}")

    df = None

    for enc in encodings:
        try:
            df = pd.read_csv(file, encoding=enc)
            break
        except:
            pass

    if df is None:
        print(f"❌ Could not read {file.name}")
        continue

    # Track where each record came from
    df["Source File"] = file.name

    all_datasets.append(df)

# Merge without changing any columns
master_df = pd.concat(all_datasets, ignore_index=True, sort=False)

print("\nMerge Completed")
print("Rows    :", master_df.shape[0])
print("Columns :", master_df.shape[1])

master_df.to_csv(
    OUTPUT_PATH / "university_raw_data.csv",
    index=False
)

print("\nRaw master dataset saved successfully.")
```
