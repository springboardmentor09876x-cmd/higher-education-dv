# 📊 Module 1 – Dataset Preparation

## Objective

Prepare a standardized raw dataset for the Higher Education Dashboard project from the integrated master dataset.

### Input
- master_dataset.csv

### Output
- university_raw_data.csv

### Tasks
- Select dashboard-relevant columns
- Recover static university information
- Recover geographical information
- Validate dataset quality
- Export the Module 1 deliverable

```python
from pathlib import Path

import numpy as np
import pandas as pd

pd.set_option("display.max_columns", None)
```

```python
# ==========================================
# Project Paths
# ==========================================

PROJECT_ROOT = Path.cwd().parent.parent

INPUT_PATH = (
    PROJECT_ROOT
    / "datasets"
    / "final"
    / "intermediate"
    / "master_dataset.csv"
)

OUTPUT_PATH = (
    PROJECT_ROOT
    / "datasets"
    / "final"
    / "Module_1_Deliverables"
)

OUTPUT_FILE = (
    OUTPUT_PATH
    / "university_raw_data.csv"
)
```

```python
df = pd.read_csv(
    INPUT_PATH,
    low_memory=False
)

print("=" * 60)
print("MASTER DATASET LOADED")
print("=" * 60)

print(f"Rows    : {len(df):,}")
print(f"Columns : {df.shape[1]}")

df.head()
```

```python
# ==========================================
# Initial Dataset Quality
# ==========================================

missing = df.isna().sum().sum()

cells = df.shape[0] * df.shape[1]

print("=" * 60)
print("INITIAL DATA QUALITY")
print("=" * 60)

print(f"Missing Values : {missing:,}")

print(
    f"Completeness : {100-(missing/cells*100):.2f}%"
)
```

```python
# ==========================================
# Prepare Dashboard Schema
# ==========================================

module1_df = df.copy()

# ------------------------------------------
# Create Unified Columns
# ------------------------------------------

module1_df["Overall Score"] = (
    module1_df["Overall Score_QS"]
    .fillna(module1_df["Overall Score_THE"])
)

module1_df["Students to Staff Ratio"] = (
    module1_df["Students to Staff Ratio_QS"]
    .fillna(module1_df["Students to Staff Ratio_THE"])
)

module1_df["International Students"] = (
    module1_df["International Students_QS"]
    .fillna(module1_df["International Students_THE"])
)

print("Unified columns created.")
```

```python
# ==========================================
# Remove Unnecessary Columns
# ==========================================

drop_columns = [

    # Duplicate QS/THE columns
    "Overall Score_QS",
    "Overall Score_THE",

    "Students to Staff Ratio_QS",
    "Students to Staff Ratio_THE",

    "International Students_QS",
    "International Students_THE",

    # Metadata not required in final schema
    "University Size",
    "University Focus",
    "Research",
    "University Status",
    "Female to Male Ratio",

    # Research duplicate
    "University Type_Research",

    # Technical columns
    "link",
    "logo",
    "Source File",
    "ScoreScaled",

    # Rank-only columns
    "Academic Reputation Rank",
    "Employer Reputation Rank",
    "Faculty Student Rank",
    "Citations per Faculty Rank",
    "International Faculty Rank",
    "International Students Rank",
    "International Research Network Rank",
    "Employment Outcomes Rank",
    "Sustainability Rank",

    # Duplicate rank columns
    "Rank_QS",
    "Rank_THE"
]

drop_columns = [
    c for c in drop_columns
    if c in module1_df.columns
]

module1_df = module1_df.drop(columns=drop_columns)

print(f"Columns Remaining : {module1_df.shape[1]}")
```

```python
# ==========================================
# University-wise Static Recovery
# ==========================================

static_columns = [

    "Country",
    "Country Code",
    "City",
    "Region",
    "Continent",
    "University Type",
    "Subject Area",
    "University Age Band"

]

for column in static_columns:

    if column not in module1_df.columns:
        continue

    module1_df[column] = (

        module1_df
        .groupby("University")[column]
        .transform(lambda x: x.ffill().bfill())

    )

print("University recovery completed.")
```

```python
# ==========================================
# Country-wise Recovery
# ==========================================

country_columns = [

    "Country Code",
    "Region",
    "Continent",
    "University Type",
    "Subject Area"

]

for column in country_columns:

    if column not in module1_df.columns:
        continue

    lookup = (

        module1_df
        .groupby("Country")[column]
        .agg(

            lambda x:

            x.mode().iloc[0]

            if not x.mode().empty

            else np.nan

        )

    )

    module1_df[column] = (

        module1_df[column]

        .fillna(

            module1_df["Country"].map(lookup)

        )

    )

print("Country recovery completed.")
```
