# Module 2 – Data Cleaning and Preprocessing

## Objective

The objective of this notebook is to clean and preprocess the integrated higher education master dataset created in Module 1.

The cleaning process includes:

- Loading the master dataset
- Inspecting dataset quality

- Removing duplicate information
- Standardizing textual data
- Recovering static university information

### Input

```python
section("DATA PREPROCESSING")

# ==========================================
# Missing Values Before Merge
# ==========================================

before_type = df["University Type"].isna().sum()
before_ratio = (
    df["Students to Staff Ratio_QS"]
    .combine_first(df["Students to Staff Ratio_THE"])
    .isna()
    .sum()
)
before_students = (
    df["International Students_QS"]
    .combine_first(df["International Students_THE"])
    .isna()
    .sum()
)
before_score = (
    df["Overall Score_QS"]
    .combine_first(df["Overall Score_THE"])
    .isna()
    .sum()
)

# ==========================================
# Merge Duplicate Columns
# ==========================================

df["University Type"] = (
    df["University Type"]
    .combine_first(df["University Type_Research"])
)

df["Students to Staff Ratio"] = (
    df["Students to Staff Ratio_QS"]
    .combine_first(df["Students to Staff Ratio_THE"])
)

df["International Students"] = (
    df["International Students_QS"]
    .combine_first(df["International Students_THE"])
)

df["Overall Score"] = (
    df["Overall Score_QS"]
    .combine_first(df["Overall Score_THE"])
)

# ==========================================
# Missing Values After Merge
# ==========================================

after_type = df["University Type"].isna().sum()
after_ratio = df["Students to Staff Ratio"].isna().sum()
after_students = df["International Students"].isna().sum()
after_score = df["Overall Score"].isna().sum()

# ==========================================
# Remove Duplicate Columns
# ==========================================

df.drop(
    columns=[
        "University Type_Research",
        "Students to Staff Ratio_QS",
        "Students to Staff Ratio_THE",
        "International Students_QS",
        "International Students_THE",
        "Overall Score_QS",
        "Overall Score_THE",
        "logo",
        "link"
    ],
    inplace=True,
    errors="ignore"
)

# ==========================================
# Standardize Text Columns
# ==========================================

text_columns = [

    "University",
    "Country",
    "Country Code",
    "City",
    "Region",

    "University Type",
    "University Size",
    "University Focus",
    "Research",
    "University Status",
    "University Age Band",

    "Subject Area"

]

for column in text_columns:

    if column in df.columns:

        df[column] = (
            df[column]
            .astype("string")
            .str.strip()
            .replace("", pd.NA)
        )

# ==========================================
# Remove Duplicate Records
# ==========================================

before_rows = len(df)

df.drop_duplicates(inplace=True)

after_rows = len(df)

# ==========================================
# Summary
# ==========================================

print("\nData Recovery Summary")
print("-" * 45)

print(f"University Type              : {before_type} → {after_type}")
print(f"Students to Staff Ratio      : {before_ratio} → {after_ratio}")
print(f"International Students       : {before_students} → {after_students}")
print(f"Overall Score                : {before_score} → {after_score}")

print("\nDuplicate Rows Removed :", before_rows - after_rows)
print(f"Remaining Rows         : {after_rows}")
```
