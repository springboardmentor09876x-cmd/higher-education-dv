```python
import pandas as pd
import numpy as np
```

```python
# ==========================
# Load Dataset
# ==========================
df = pd.read_csv("University_Cleaned.csv")
```

```python
# ==========================
# Helper Function
# ==========================
def normalize(series):
    """
    Min-Max Normalization (0-100)
    """
    series = pd.to_numeric(series, errors='coerce')

    if series.max() == series.min():
        return pd.Series(50, index=series.index)

    return ((series - series.min()) /
            (series.max() - series.min())) * 100
```

```python
# ==========================
# KPI 1: Global Ranking Score
# ==========================
# Lower rank is better, so invert rankings

df["QS_Rank_Num"] = pd.to_numeric(df["QS_Rank"], errors="coerce")
df["THE_Rank_Num"] = pd.to_numeric(df["THE_Rank"], errors="coerce")

df["Global_Ranking_Score"] = (
    0.4 * normalize(df["QS_Rank_Num"].max() - df["QS_Rank_Num"]) +
    0.3 * normalize(df["THE_Rank_Num"].max() - df["THE_Rank_Num"]) +
    0.3 * normalize(df["Overall Score"])
)
```

```python
# ==========================
# KPI 2: Research Impact Score
# ==========================
df["Research_Impact_Score"] = (
    0.5 * normalize(df["Citation_Score"]) +
    0.5 * normalize(df["Research Quality"])
)
```

```python
# ==========================
# KPI 3: Faculty-to-Student Ratio
# ==========================
df["Faculty_to_Student_Ratio_KPI"] = pd.to_numeric(
    df["Student_Faculty_Ratio"],
    errors="coerce"
)
```

```python
# ==========================
# KPI 4: International Student Percentage
# ==========================
df["International_Student_Percentage"] = pd.to_numeric(
    df["International_Students (%)"],
    errors="coerce"
)
```

```python
# ==========================
# KPI 5: Academic Reputation Score
# ==========================
df["Academic_Reputation_Score_KPI"] = pd.to_numeric(
    df["Academic Reputation Score"],
    errors="coerce"
)
```

```python
# ==========================
# KPI 6: Research Productivity Index
# ==========================
df["Research_Productivity_Index"] = (
    0.6 * normalize(df["Publication_Count"]) +
    0.4 * normalize(df["H_Index"])
)
```

```python
# ==========================
# KPI 7: Academic Excellence Index
# ==========================
df["Academic_Excellence_Index"] = (
    0.4 * normalize(df["Academic Reputation Score"]) +
    0.3 * normalize(df["Employer Reputation Score"]) +
    0.3 * normalize(df["Faculty Student Score"])
)
```

```python
# ==========================
# KPI 8: Research Efficiency Score
# ==========================
df["Research_Efficiency_Score"] = (
    df["Publication_Count"] /
    df["Faculty_Count"]
)
```

```python
# ==========================
# KPI 9: Citation Efficiency Score
# ==========================
df["Citation_Efficiency_Score"] = (
    df["Citations_Count"] /
    df["Faculty_Count"]
)
```

```python
# ==========================
# KPI 10: Student Diversity Index
# ==========================
df["Gender_Balance"] = (
    100 - abs(df["male_percentage"] - df["female_percentage"])
)

df["Student_Diversity_Index"] = (
    0.6 * normalize(df["International_Students (%)"]) +
    0.4 * normalize(df["Gender_Balance"])
)
```

```python
# ==========================
# KPI 11: Employability Excellence Score
# ==========================
df["Employability_Excellence"] = (
    0.6 * normalize(df["Employer Reputation Score"]) +
    0.4 * normalize(
        df["Employability_Rank"].max() -
        df["Employability_Rank"]
    )
)
```

```python
# ==========================
# KPI 12: Global Engagement Index
# ==========================
df["Global_Engagement_Index"] = (
    0.5 * normalize(df["International_Students (%)"]) +
    0.5 * normalize(df["International Outlook"])
)
```

```python
# ==========================
# KPI 13: Affordability Score
# ==========================
df["Affordability_Score"] = (
    normalize(
        df["Tuition_Fee_USD"].max() -
        df["Tuition_Fee_USD"]
    )
)
```

```python
# ==========================
# KPI 14: Student Success Index
# ==========================
df["Student_Success_Index"] = (
    0.5 * normalize(df["Graduation_Rate (%)"]) +
    0.5 * normalize(df["Employment Outcomes Score"])
)
```

```python
# ==========================
# KPI 15: Value-for-Money Index
# ==========================
df["Value_For_Money_Index"] = (
    df["Overall Score"] /
    df["Tuition_Fee_USD"]
)
```

```python
# ==========================
# KPI 16: Research Leadership Index
# ==========================
df["Research_Leadership_Index"] = (
    0.35 * normalize(df["Research Quality"]) +
    0.25 * normalize(df["Citation_Score"]) +
    0.20 * normalize(df["Publication_Count"]) +
    0.20 * normalize(df["H_Index"])
)
```

```python
# ==========================
# KPI 17: University Competitiveness Index
# ==========================
df["University_Competitiveness_Index"] = (
    0.4 * normalize(df["Overall Score"]) +
    0.3 * normalize(
        df["Acceptance_Rate (%)"].max() -
        df["Acceptance_Rate (%)"]
    ) +
    0.3 * normalize(df["Academic Reputation Score"])
)
```

```python
# ==========================
# KPI 18: Future Potential Index
# ==========================
df["Future_Potential_Index"] = (
    0.4 * normalize(df["Research Environment"]) +
    0.3 * normalize(df["Industry Impact"]) +
    0.3 * normalize(df["International Research Network Score"])
)
```

```python
# ==========================
# Tableau Final Dataset
# ==========================

tableau_columns = [

    # University Information
    "University",
    "Country",
    "Region",
    "City_Final",

    # Core Rankings
    "QS_Rank",
    "THE_Rank",
    "CWUR_Rank",
    "Overall Score",

    # Base Metrics
    "Total_Students",
    "Faculty_Count",
    "Student_Faculty_Ratio",
    "International_Students (%)",
    "Academic Reputation Score",
    "Employer Reputation Score",

    # Research Metrics
    "Publication_Count",
    "H_Index",
    "Citations_Count",
    "Citation_Score",
    "Research Quality",

    # Outcome Metrics
    "Graduation_Rate (%)",
    "Acceptance_Rate (%)",
    "Employment Outcomes Score",
    "Tuition_Fee_USD",

    # ==========================
    # Required Milestone KPIs
    # ==========================
    "Global_Ranking_Score",
    "Research_Impact_Score",
    "Faculty_to_Student_Ratio_KPI",
    "International_Student_Percentage",
    "Academic_Reputation_Score_KPI",
    "Research_Productivity_Index",

    # ==========================
    # Advanced EduVision KPIs
    # ==========================
    "Academic_Excellence_Index",
    "Research_Efficiency_Score",
    "Citation_Efficiency_Score",
    "Student_Diversity_Index",
    "Employability_Excellence",
    "Global_Engagement_Index",
    "Affordability_Score",
    "Student_Success_Index",
    "Value_For_Money_Index",
    "Research_Leadership_Index",
    "University_Competitiveness_Index",
    "Future_Potential_Index"
]
```

```python
# Keep only existing columns
tableau_columns = [
    col for col in tableau_columns
    if col in df.columns
]

final_df = df[tableau_columns]
```

```python
# ==========================
# Save Files
# ==========================

final_df.to_excel(
    "university_final_dataset.xlsx",
    index=False
)

final_df.to_csv(
    "university_final_dataset.csv",
    index=False
)

print("KPI Engineering Completed")
print("university_final_dataset.xlsx generated")
print("university_final_dataset.csv generated")
print(f"Rows: {len(final_df)}")
print(f"Columns: {len(final_df.columns)}")
```
