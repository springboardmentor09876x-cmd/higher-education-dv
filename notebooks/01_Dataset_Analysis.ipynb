```python
#Standardize University Names

import re

def clean_university_name(name):
    if pd.isna(name):
        return name

    name = str(name).lower().strip()

    # Remove text inside brackets
    name = re.sub(r'\(.*?\)', '', name)

    # Remove punctuation
    name = re.sub(r'[^\w\s]', '', name)

    # Replace common abbreviations
    name = re.sub(r'\buniv\b', 'university', name)
    name = re.sub(r'\binst\b', 'institute', name)
    name = re.sub(r'\btech\b', 'technology', name)

    # Remove extra spaces
    name = re.sub(r'\s+', ' ', name).strip()

    return name
```

```python
for df in datasets.values():
    df["University_Name_Clean"] = df["University_Name"].apply(clean_university_name)
```
