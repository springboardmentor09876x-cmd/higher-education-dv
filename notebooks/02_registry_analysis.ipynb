```python
import pandas as pd

pd.set_option("display.max_rows", 100)
pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", None)
```

```python
registry = pd.read_csv("../data/interim/registry/university_registry.csv")

print("Shape :", registry.shape)

registry.head()
```

```python
print(registry.columns.tolist())
```

```python
print("Duplicate normalized names :")

print(
    registry["normalized_name"]
    .duplicated()
    .sum()
)
```

```python
duplicates = registry[
    registry["normalized_name"].duplicated(keep=False)
].sort_values("normalized_name")

duplicates
```

```python
print("Unique university names :")

print(
    registry["university_name"]
    .nunique()
)

print()

print("Unique normalized names :")

print(
    registry["normalized_name"]
    .nunique()
)
```

```python
registry["word_count"] = (
    registry["university_name"]
    .str.split()
    .str.len()
)

registry.sort_values(
    "word_count",
    ascending=False
)[
    [
        "university_name",
        "normalized_name",
        "country",
        "word_count",
    ]
].head(50)
```
