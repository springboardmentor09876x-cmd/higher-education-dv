```python
from pathlib import Path
import pandas as pd
import os
import sys

# Go up one level if your notebook is in a subfolder (like 'notebooks')
# Adjust ".." to your actual path if your structure is deeper!
sys.path.append(os.path.abspath(".."))

from scripts.data_collection import load_raw_datasets, main

from scripts.data_collection import (
    load_raw_datasets,
    main
)
```

```python
RAW_PATH = Path("../datasets/raw")
OUTPUT_PATH = Path("../datasets/final")

print("Raw Folder :", RAW_PATH)
print("Output Folder :", OUTPUT_PATH)
```

```python
csv_files = sorted(RAW_PATH.rglob("*.csv"))

print(f"Datasets Found : {len(csv_files)}\n")

for file in csv_files:
    print(file.name)
```

```python
datasets = load_raw_datasets()

print(f"Loaded {len(datasets)} datasets")
```

```python
summary = pd.DataFrame({
    "Dataset": [f.name for f in csv_files],
    "Rows": [df.shape[0] for df in datasets],
    "Columns": [df.shape[1] for df in datasets]
})

summary
```

```python
datasets[0].head()
```

```python
all_data = load_raw_datasets()

print(f"Total datasets processed : {len(all_data)}")

master_df = pd.concat(all_data, ignore_index=True, sort=False)

print("\nMaster Dataset Created Successfully")
print(f"Rows    : {master_df.shape[0]}")
print(f"Columns : {master_df.shape[1]}")

master_df.head()
```

```python
OUTPUT_PATH.mkdir(parents=True, exist_ok=True)

output_file = OUTPUT_PATH / "master_dataset.csv"

master_df.to_csv(output_file, index=False)

print(f"Saved Successfully : {output_file}")
```

```python
master_df.info()
```

```python
missing = (
    master_df.isnull()
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

missing.columns = ["Column", "Missing Values"]

missing
```

```python
print("=" * 60)
print("MASTER DATASET SUMMARY")
print("=" * 60)

print(f"Rows              : {master_df.shape[0]}")
print(f"Columns           : {master_df.shape[1]}")
print(f"Duplicate Rows    : {master_df.duplicated().sum()}")

print("\nColumns")

for col in master_df.columns:
    print(f"• {col}")
```

```python
print("=" * 60)
print("COLUMN COMPLETENESS")
print("=" * 60)

completeness = pd.DataFrame({
    "Column": master_df.columns,
    "Non-Null Values": master_df.notnull().sum().values,
    "Missing Values": master_df.isnull().sum().values,
    "Completeness (%)": (
        master_df.notnull().mean() * 100
    ).round(2).values
})

completeness.sort_values(
    by="Completeness (%)",
    ascending=False
)
```
