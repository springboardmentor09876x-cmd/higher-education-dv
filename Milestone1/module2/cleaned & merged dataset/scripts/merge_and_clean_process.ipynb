```python
import pandas as pd

# Read Excel file (.xls)
df1 = pd.read_excel("Top 1000 Universities Worldwide.xlsx")

# Read CSV files
df2 = pd.read_csv("top_100_universities_dataset.csv")
df3 = pd.read_excel("new_world_university_rankings_2026.xlsx")
df4 = pd.read_excel("openalex_institutions_pk_core.xlsx")
df5 = pd.read_csv("dataset_for_missing_columns.csv")

print("All datasets loaded successfully!")
```

```python
print("Dataset 1")
print(df1.columns.tolist())

print("\nDataset 2")
print(df2.columns.tolist())

print("\nDataset 3")
print(df3.columns.tolist())

print("\nDataset 4")
print(df4.columns.tolist())

print("\nDataset 5")
print(df5.columns.tolist())
```

```python
df1.rename(columns={"University_Name": "university_name"}, inplace=True)
df2.rename(columns={"University_Name": "university_name"}, inplace=True)
```

```python
for df in [df1, df2, df3, df4]:
    df["university_name"] = (
        df["university_name"]
        .astype(str)
        .str.strip()
        .str.lower()
    )
```

```python
for df in [df1, df2, df3, df4]:
    df.drop_duplicates(subset="university_name", inplace=True)
```

```python
merged = df1.merge(df2, on="university_name", how="outer")
merged = merged.merge(df3, on="university_name", how="outer")
merged = merged.merge(df4, on="university_name", how="outer")
```

```python
import pandas as pd

df = pd.read_csv("University_Final_Data_mo.csv")
```

```python
print(df.shape)
print(df.info())
print(df.head())
```

```python
print(df.columns.tolist())
```

```python
df["country"] = (
    df["country_x"]
    .combine_first(df["country_y"])
    .combine_first(df["Country"])
)

df.drop(columns=["country_x", "country_y", "Country"], inplace=True)
```

```python
df["region"] = df["region_x"].combine_first(df["region_y"])

df.drop(columns=["region_x", "region_y"], inplace=True)
```

```python
df["university_type"] = (
    df["university_type_x"]
    .combine_first(df["university_type_y"])
    .combine_first(df["University_Type"])
)

df.drop(
    columns=[
        "university_type_x",
        "university_type_y",
        "University_Type"
    ],
    inplace=True
)
```

```python
df["total_students"] = (
    df["total_students_x"]
    .combine_first(df["total_students_y"])
    .combine_first(df["Total_Students"])
)

df.drop(
    columns=[
        "total_students_x",
        "total_students_y",
        "Total_Students"
    ],
    inplace=True
)
```

```python
df["world_rank"] = (
    df["world_rank"]
    .combine_first(df["World_Rank"])
)

df.drop(columns=["World_Rank"], inplace=True)
```

```python
df["national_rank"] = (
    df["national_rank"]
    .combine_first(df["National_Rank"])
)

df.drop(columns=["National_Rank"], inplace=True)
```

```python
df["faculty_count"] = (
    df["faculty_count"]
    .combine_first(df["Faculty_Count"])
)

df.drop(columns=["Faculty_Count"], inplace=True)
```

```python
drop_cols = [
    "works_api_url",
    "homepage_url",
    "updated_date",
    "created_date",
    "ids.openalex",
    "ids.ror",
    "ids.wikipedia",
    "ids.wikidata",
]

df.drop(columns=drop_cols, inplace=True, errors="ignore")
```

```python
print(df.shape)
print(df.columns.tolist())
```

```python
required_columns = [
    "university_name",
    "country",
    "region",
    "university_type",
    "qs_rank_2026",
    "the_rank_2026",
    "arwu_rank_2025",
    "qs_score",
    "the_score",
    "arwu_score",
    "qs_academic_rep",
    "qs_employer_rep",
    "qs_faculty_student",
    "qs_citations",
    "qs_intl_faculty",
    "qs_intl_students",
    "the_teaching",
    "the_research_env",
    "the_research_quality",
    "the_industry",
    "the_intl_outlook",
    "founded",
    "total_students",
    "intl_students_pct",
    "nobel_laureates",
    "fields_medalists",
    "ranked_in_all_3",
    "avg_rank_all3",
    "university_age_2026"
]

missing = [col for col in required_columns if col not in df.columns]

print("Missing Columns:", missing)
print("Available:", len(required_columns) - len(missing), "/", len(required_columns))
```

```python
df.to_csv("University_Cleaned.csv", index=False)
print("University_Cleaned.csv saved successfully!")
```
