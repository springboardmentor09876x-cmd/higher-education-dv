```python
import pandas as pd
import numpy as np

print("Pandas version:", pd.__version__)
print("NumPy version:", np.__version__)
```

```python
qs = pd.read_csv("Data2/QS2024.csv",encoding="latin1")
the = pd.read_csv("Data2/QS2025.csv",encoding="latin1")
```

```python
qs.head()
```

```python
the.head()
```

```python
the.rename(columns={
    "Institution Name":"Institution_Name"
}, inplace=True)
```

```python
times = pd.read_csv("Data2/timesData.csv")

shanghai = pd.read_csv("Data2/shanghaiData.csv")
```

```python
times.head()

```

```python
shanghai.head()
```

```python
qs.rename(columns={
    "Institution_Name": "university_name"
}, inplace=True)

the.rename(columns={
    "Institution Name": "university_name"
}, inplace=True)
```

```python
the.head()
```

```python
qs.head()
```

```python
for df in [qs, the, times, shanghai]:
    df["university_name"] = (
        df["university_name"]
        .str.strip()
        .str.lower()
    )
```

```python
merged = pd.merge(
    qs,
    the,
    on="university_name",
    how="outer"
)
```

```python
merged = pd.merge(
    merged,
    times,
    on="university_name",
    how="outer"
)
```

```python
merged = pd.merge(
    merged,
    shanghai,
    on="university_name",
    how="outer"
)
```
